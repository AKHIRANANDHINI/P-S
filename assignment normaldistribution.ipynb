```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Parameters
mean = 30
std_dev = 7
num_samples = 1000

# Generate data points from a normal distribution
data = np.random.normal(mean, std_dev, num_samples)

# Plot histogram
plt.figure(figsize=(10, 5))
plt.hist(data, bins=30, density=True, alpha=0.6, color='blue', edgecolor='black', label='Histogram')

# Plot PDF
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 1000)
pdf = stats.norm.pdf(x, mean, std_dev)
plt.plot(x, pdf, 'r', label='PDF', linewidth=2)

# Labels and title
plt.xlabel('Data Values')
plt.ylabel('Probability Density')
plt.title('Normal Distribution: Histogram & PDF')
plt.legend()
plt.grid()

# Show plot
plt.show()
```
